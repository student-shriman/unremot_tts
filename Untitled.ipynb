```python
## Loading and processing the dataset ..
import pandas as pd
pd.set_option("display.max_colwidth",None)

food = pd.read_csv(r'data/Reviews.csv')
lounge = pd.read_csv(r'data/lounge.csv')
others =  pd.read_csv(r'data/others_test.csv')

food = food.iloc[:2000,:]
lounge = lounge.iloc[:2000, :] 
others = others.iloc[:2000, :]

##  Getting food df
food_category = []
for i in range(2000):
    food_category.append('food')
food['food_category']=food_category
food = food[['Text', 'food_category']]

##  Getting lounge df
lounge_category = []
for i in range(2000):
    lounge_category.append('lounge')
lounge['lounge_category']=lounge_category
lounge = lounge[['content', 'lounge_category']]

##  Getting others df
others_category = []
for i in range(2000):
    others_category.append('others')
others['others_category']=others_category

print ('No. of reviews for food category - ', len(food.loc[:]))
print ('------------------------------------------')
print ('No. of reviews for lounge category - ', len(lounge.loc[:]))
print ('------------------------------------------')
print ('No. of reviews for others category - ', len(others.loc[:]))
```

```python
## loading teh structured data ..
food = pd.read_csv(r'food.csv')
lounge = pd.read_csv(r'lounge.csv')
others = pd.read_csv(r'others.csv')

print(food.columns)
print(lounge.columns)
print(others.columns)
```

```python
##  Splitting train test sets for category classification ..
food_train = food.iloc[0:1700,:]
food_test = food.iloc[1700:2000,:]

lounge_train = lounge.iloc[0:1700,:]
lounge_test = lounge.iloc[1700:2000,:]

others_train = others.iloc[0:1700,:][['text', 'category']]
others_test = others.iloc[1700:2000,:][['text', 'category']]

print ('No. of reviews for food train - ', len(food_train.loc[:]))
print ('No. of reviews for food test - ', len(food_test.loc[:]))
print ('------------------------------------------')

print ('No. of reviews for lounge train - ', len(lounge_train.loc[:]))
print ('No. of reviews for lounge test - ', len(lounge_test.loc[:]))
print ('------------------------------------------')

print ('No. of reviews for others train - ', len(others_train.loc[:]))
print ('No. of reviews for others test - ', len(others_test.loc[:]))
```

```python
##  Removing un-wanted columns  ..
train_df = pd.concat([food_train, lounge_train, others_train], axis=0)
test_df = pd.concat([food_test, lounge_test, others_test], axis=0)
train_df=train_df[['text', 'category']]
test_df=test_df[['text', 'category']]
```

##  Getting Sentiment analysis using vadersentiments library ..

```python
import facebook_scraper
from facebook_scraper import get_posts
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

##  Defining a function for getting sentiments ..
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    if score['compound'] >= 0.05:
        result = 'pos'
    elif score['compound'] > -0.05 and score['compound'] < 0.05:
        result = 'neu'
    else:
        result = 'neg'
    print (score, result, score[result], sep='    ')
    print ('----------------')
    return result,  score[result]
```

```python
sentiments = []
confidences = []

for i in train_df.itertuples():
    label, conf = sentiment_analyzer_scores(i[1])
    if label == 'pos':
        sentiment = 'Positive'
    elif label == 'neu':
        sentiment = 'Neutral'
    elif label == 'neg':
        sentiment = 'Negative'
    
    sentiments.append(sentiment) 
    confidences.append(conf)
```
